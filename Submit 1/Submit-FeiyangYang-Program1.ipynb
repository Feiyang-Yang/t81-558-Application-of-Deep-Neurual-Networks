```python

# coding: utf-8

# In[3]:




# In[45]:

import os
import sklearn
from sklearn.cross_validation import KFold
import pandas as pd
import numpy as np
from scipy.stats import zscore

path = "./data/"

#Question 1
filename_read = os.path.join(path,"auto-mpg.csv")
df = pd.read_csv(filename_read,na_values=['NA','?'])

df = df['weight']

average = df.mean()
    
print('***Question 1***')
print('Mean weight is: {}'.format(average))

#Question 2
#Replace NA of horsepower with Median
filename_read = os.path.join(path,"auto-mpg.csv")
df = pd.read_csv(filename_read,na_values=['NA','?'])
med = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(med)

    #check if there still exists NA in horsepower column
if pd.isnull(df['horsepower']).values.any() == True:
    print('There still exist NA in horsepower')

#transform mpg, displace- ment, horsepower, weight &acceleration col to Z-Scores
df['mpg'] = zscore(df['mpg'])
df['displacement'] = zscore(df['displacement'])
df['horsepower'] = zscore(df['horsepower'])
df['weight'] = zscore(df['weight'])
df['acceleration'] = zscore(df['acceleration'])
df

#Transform to and save the csv file
filename_write = os.path.join(path,"Submit_FeiyangYang_Pro1_Q2.csv")
df.to_csv(filename_write,index=False)

#Get average and standard deviation of MPG then print out
MPGaverage = df['mpg'].mean()
MPGstan_devia = df['mpg'].std()
print('***Question 2***')
print('MPG mean: {}'.format(MPGaverage))
print('MPG standard deviation: {}'.format(MPGstan_devia))



#Question 3
#Calculate efficiency item
filename_read = os.path.join(path,"auto-mpg.csv")
df = pd.read_csv(filename_read,na_values=['NA','?'])
df.insert(1,'efficiency',(df['horsepower']/df['displacement']).astype(float))
df

#Create a new dataset that has only the car name and efficiency
col_name = df['name']
col_efficiency = df['efficiency']
result = pd.concat([col_name,col_efficiency],axis=1)
result

#Sort this dataset by efficency
result = result.sort_values(by='efficiency',ascending = False)
result
#print('***Question 3***')
#print('Most efficient is:'result[0,0])
      
#Transform to and save the csv file
filename_write = os.path.join(path,"Submit_FeiyangYang_Pro1_Q3.csv")
result.to_csv(filename_write,index=False)

#print the most efficient one
print('***Question 3***')
print("The most efficient is: {}".format(result['name'].iloc[0]))

#Question 4
#Seperate into two sets
filename_read = os.path.join(path,"auto-mpg.csv")
df = pd.read_csv(filename_read,na_values=['NA','?'])
df = df.reindex(np.random.permutation(df.index)) # Usually a good idea to shuffle
mask = np.random.rand(len(df)) < 0.8
trainDF = pd.DataFrame(df[mask])
validationDF = pd.DataFrame(df[~mask])

#Print out the size of each set
print('***Question 4***')
print("Training DF: {}".format(len(trainDF)))
print("Validation DF: {}".format(len(validationDF)))

#Name the column and insert it into the csv file with TV
trainDF.insert(1,'set','T')
validationDF.insert(1,'set','V')
result=pd.concat([trainDF,validationDF])

#Transform to and save the csv file
filename_write = os.path.join(path,"Submit_FeiyangYang_Pro1_Q4.csv")
result.to_csv(filename_write,index=False)


#Question 5
import os
from sklearn.model_selection import KFold
import pandas as pd
import numpy as np
path = "./data/"

#Seperate into five folds preparation
filename_read = os.path.join(path,"auto-mpg.csv")
df = pd.read_csv(filename_read,na_values=['NA','?'])
df = df.reindex(np.random.permutation(df.index))
kf = KFold(5)

#for loop setted to print and add to a new dataframe
print('***Question 5***')
fold = 1
result = [] # create a new dataframe called 'result'
for train_index, validate_index in kf.split(df):        
    trainDF = pd.DataFrame(df.ix[train_index,:])
    validateDF = pd.DataFrame(df.ix[validate_index])
    print("Fold #{}, Training Size: {}, Validation Size: {}".format(fold,len(trainDF),len(validateDF)))
    fold+=1
    #Name the column and insert it into the csv file with TV
    trainDF.insert(0,'set','T')  
    validateDF.insert(0,'set','V')  
    trainDF.insert(1,'iteration',(fold))  
    validateDF.insert(1,'iteration',(fold)) 
    #result=pd.concat([trainDF,validateDF])  
    #result.append(result) #JH - It is overwritting your existing array each time through the loop, so       #did not get     all of the data loaded.
    result.append(trainDF) #JH - just add each of these to your list
    result.append(validateDF) #JH
result = pd.concat(result) #JH - now concat all 10 of the items added to the list together, this gives #                            #you the complete dataset
result 

#Transform to and save the csv file
filename_write = os.path.join(path,"Submit_FeiyangYang_Pro1_Q5.csv")
result.to_csv(filename_write,index=False)







# In[ ]:




```
